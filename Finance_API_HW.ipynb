# Task 1: Fetch Basic Stock Data

```python
import json
import pandas as pd
import json
import requests
```

```python
apikey = ""

headers = {
    'x-api-key': apikey
    }
```

```python
stocks=input("Enter the stock(s) ticker names (max 10): ")
```

```python
querystring = {"symbols": stocks}

url = "https://yfapi.net/v6/finance/quote"

response = requests.get(url, headers=headers, params=querystring)

# Convert response to DataFrame
df = pd.DataFrame(response.json()['quoteResponse']['result'])
df.head()
```

```python
for i in range(len(df)):
  print("Stock Ticker: %s, Company: %s, Current Market Price: $%.2f" % (df['symbol'][i], df['shortName'][i], df['regularMarketPrice'][i]))
```

# Task 2.1: Fetch Additional Data Using Modules

## Choose a module

```python
def get_a_module(stock, module):
  querystring = {"modules":module}

  url = "https://yfapi.net/v11/finance/quoteSummary/"
  url = url + stock

  response = requests.get(url, headers=headers, params=querystring)
  # Convert response to DataFrame
  df = pd.DataFrame(response.json()['quoteSummary']['result'][0][module])
  return df
```

```python
stock=input("Enter a single stock ticker name: ")
```

```python
summary_detail_df = get_a_module(stock, "summaryDetail")
financial_df = get_a_module(stock, "financialData")
```

```python
fiftyTwoWeekHigh = summary_detail_df['fiftyTwoWeekHigh']['fmt']
fiftyTwoWeekLow = summary_detail_df['fiftyTwoWeekLow']['fmt']
returnOnAssets = financial_df['returnOnAssets']['fmt']
```

```python
print("Stock: " + stock)
print("Fifty Two Week High: $" + fiftyTwoWeekHigh)
print("Fifty Two Week LowL: $" + fiftyTwoWeekLow)
print("Return On Assets (ROA): " + returnOnAssets)
```

```python
region = "US"

querystring = {}

url = "https://yfapi.net/v1/finance/trending/"
url = url + region

response = requests.get(url, headers=headers, params=querystring)
df = pd.DataFrame(response.json()['finance'])
```

```python
df = pd.DataFrame(response.json()['finance']['result'])
```

```python
df_new = pd.DataFrame(df['quotes'][0])
df_new.head(20)
```

```python
def get_a_quote(symbols):
  querystring = {"symbols": symbols}

  url = "https://yfapi.net/v6/finance/quote"

  response = requests.get(url, headers=headers, params=querystring)
  return response.json()
```

```python
trending_stocks = []
for i in range(len(df_new)):
  trending_stocks.append(df_new['symbol'][i])
symbols = ""
count = 0
for stock in trending_stocks:
  if count < 9:
    symbols = symbols + stock + ","
    count += 1
  else:
    symbols = symbols + stock
    data = get_a_quote(symbols)
    df = pd.DataFrame(data['quoteResponse']['result'])
    for i in range(len(df)):
      print("""
      Trending Stock Name: %s,
      Ticker: %s,
      Current Price: $%.2f,
      52 Week High: $%.2f,
      52 Week Low: $%.2f""" % (df['shortName'][i], df['symbol'][i],
                               df['regularMarketPrice'][i],
                               df['fiftyTwoWeekHigh'][i], df['fiftyTwoWeekLow'][i]))
    symbols = ""
    count = 0
    symbols = ""
```
