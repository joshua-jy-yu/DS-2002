As pyflightdata did not have data on Taylor Swift's jet, let's look at other flight data APIs. This is OpenSky.

```python
import requests
def get_flight_data():
    url = 'https://opensky-network.org/api/states/all'
    response = requests.get(url)
    data = response.json()
    return data
flight_data = get_flight_data()
```

```python
import pandas as pd
def process_flight_data(data):
    columns = ['icao24', 'callsign', 'origin_country', 'time_position', 'last_contact',
               'longitude', 'latitude', 'baro_altitude', 'on_ground', 'velocity',
               'true_track', 'vertical_rate', 'sensors', 'geo_altitude', 'squawk',
               'spi', 'position_source']
    flights_df = pd.DataFrame(data['states'], columns=columns)
    return flights_df
flight_data = get_flight_data()
flights_df = process_flight_data(flight_data)
flights_df.info()
```

The API did not have the tail number, but instead experimented with the callsign and icao24 number.

```python
flights_df.head()
```

```python
taylor_swift_df = flights_df[flights_df['icao24']== "A81B13"] # an icao24 code found online
taylor_swift_df
```
